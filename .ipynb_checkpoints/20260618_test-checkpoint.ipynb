```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create fake credit data
data = pd.DataFrame({
    'income': np.random.normal(50000, 15000, 1000),
    'age': np.random.randint(22, 65, 1000),
    'defaulted': np.random.binomial(1, 0.1, 1000)
})

# Summarize - like Stata's "summarize"
print(data.describe())

# Plot income distribution
plt.hist(data['income'], bins=30)
plt.title('Income Distribution')
plt.show()
```
